```python
users_rated = x[x].index
ratings_with_name['User-ID'].isin(users_rated)
```

```python
ratings_with_name[ratings_with_name['User-ID'].isin(users_rated)]
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(users_rated)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
#NaN ko humne 0 kar dia
```

```python
pt
```

```python
# we have 706 books which have min 50 rates each
# and 810 raters who have rated atleast 250 books
```

```python
# we will make vectors of each 706 books having 810 dimensions
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt).shape
# we found consine of angle of the above vector
# since there are 706 books so 706 coscine value with each book
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),reverse=True,key=lambda x:x[1])[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
        
    return data           
```
